# Get a list of accounts that have watched

```python
# create address to id map
addressIdMap = {}

import requests
r = requests.get("http://127.0.0.1:9002/api/v1/accounts").json()

## parse the response as to list of accounts:
## Do not include accounts with watch=False
watchedAccounts = list(filter(lambda x: x["watch"], r))
for account in watchedAccounts:
    addressIdMap[account["address"]] = account["id"]
    
accounts = list(map(lambda x: x["address"], watchedAccounts))

print(watchedAccounts)
print(accounts)
```

# Split list of accounts into chunks of 20

```python
chunks = []
for i in range(0, len(accounts), 20):
    chunks.append(accounts[i:i + 20])

print(chunks)
```
